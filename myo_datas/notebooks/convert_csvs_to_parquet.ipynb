```python
import pandas as pd
from pathlib import Path

BASE = Path('''/Users/corin44/Documents/myofitness/myo_datas''')
PROCESSED = BASE / 'data' / 'processed'

csvs = [
    'calories_cleaned.csv',
    'exercise_dataset_cleaned.csv',
    'gym_members_exercise_tracking_synthetic_data_cleaned.csv',
    'programs_detailed_cleaned.csv',
    'program_summary_cleaned.csv'
]

import importlib
for c in csvs:
    p = PROCESSED / c
    if not p.exists():
        print('missing', p)
        continue
    df = pd.read_csv(p, low_memory=False)
    out = p.with_suffix('.parquet')
    # choose available parquet engine or fallback to gzipped CSV
    engine = None
    for eng in ('pyarrow','fastparquet'):
        try:
            importlib.import_module(eng)
            engine = eng
            break
        except Exception:
            continue
    if engine is not None:
        df.to_parquet(out, engine=engine, index=False)
        print('wrote', out, 'using', engine)
    else:
        out_csvgz = p.with_suffix('.csv.gz')
        df.to_csv(out_csvgz, index=False, compression='gzip')
        print('wrote', out_csvgz, '(no parquet engine available)')

```

```python
import pandas as pd
from pathlib import Path

BASE = Path.cwd().parents[0]
RAW = BASE / "data" / "raw"
PROCESSED = BASE / "data" / "processed"
PROCESSED.mkdir(parents=True, exist_ok=True)

raw_path = RAW / "program_summary.csv"
df = pd.read_csv(raw_path, encoding='utf-8', low_memory=False)

# Basic cleaning
str_cols = df.select_dtypes(include=["object"]).columns
for c in str_cols:
    df[c] = df[c].astype(str).str.strip()

# Parse numeric columns if any
for col in df.columns:
    if df[col].dtype == object:
        # attempt numeric coercion
        df[col] = pd.to_numeric(df[col].str.replace(',',''), errors='ignore')

# Drop duplicates and empty rows
df = df.drop_duplicates()
if df.shape[1] > 0:
    df = df.dropna(how='all')

out_path = PROCESSED / 'program_summary_cleaned.csv'
df.to_csv(out_path, index=False, encoding='utf-8')

print('Saved cleaned file to', out_path)

```
